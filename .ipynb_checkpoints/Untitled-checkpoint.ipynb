```python
%pylab inline
import mdtraj as md
import numpy as np
import itertools
```

```python
J = md.load('unboundgp120.pdb')
H = md.load('H-unbound.pdb')
R = md.load('r2-unbound.pdb')
```

```python
J_CA = J.topology.select('name CA')
J_CA_pairs = list(itertools.product(J_CA, J_CA))
H_CA = H.topology.select('name CA')
H_CA_pairs = list(itertools.product(H_CA, H_CA))
R_CA = R.topology.select('name CA')
R_CA_pairs = list(itertools.product(R_CA, R_CA))
```

```python
J_distacne_map = md.compute_distances(J,atom_pairs=J_CA_pairs).reshape(J.n_residues,J.n_residues)
H_distacne_map = md.compute_distances(H,atom_pairs=H_CA_pairs).reshape(H.n_residues,H.n_residues)
R_distacne_map = md.compute_distances(R,atom_pairs=R_CA_pairs).reshape(R.n_residues,R.n_residues)
```

```python
def get_map(distance_map,top):
    max_ = np.max(distance_map)
    min_ = np.min(distance_map)
    values = np.linspace(min_,max_/2,20)
    aa = ['Ala','Arg','Asn','Asp','Cys','Gln','Glu','Gly','His','Ile','Leu','Lys','Met','Phe','Pro','Ser','Thr','Trp','Tyr','Val']
    aa = [s.upper() for s in aa ]
    aa_dict = {}
    for i,a in enumerate(aa):
        aa_dict[a] = values[i]
        
    n = top.topology.n_residues
    resi_map = np.zeros((n,n))
    
    for i,resi_i in enumerate(list(top.topology.residues)):
        for j,resi_j in enumerate(list(top.topology.residues)):
            resi_map[i,j] = aa_dict.get(str(resi_i)[:3])+ aa_dict.get(str(resi_j)[:3])
    
    map_ = np.zeros((n,n))
    for i in range(n-1):
        for j in range(i+1,n):
            map_[j,i] = distance_map[i,j]
            map_[i,j] = resi_map[i,j]
    for i in range(n):
        map_[i,i] = distance_map[i,i]
    return map_
```

```python
J_map = get_map(J_distacne_map,J)
H_map = get_map(H_distacne_map,H)
R_map = get_map(R_distacne_map,R)
```

```python
figure(figsize=(6,6))
subplot(221)
contourf(J_map)
savefig('J.png',dpi=720,bbox_inches='tight')

        
```
